```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
```

```python
Salary=pd.read_csv('Salary_Data.csv')
Salary.head()
```

```python
Salary.info()
```

```python
Salary.plot(kind='box',subplots=True)
plt.tight_layout();
```

```python
Salary.corr()
```

```python
sns.regplot(x=Salary['YearsExperience'],y=Salary['Salary'])
```

```python
formula = 'Salary ~ YearsExperience'
model = ols(formula=formula,data=Salary).fit()
print(model.summary())
```

```python
# Manual prediction for say 3 Years Experience
Salary = (25792.200199) + (9449.962321)*(3)
Salary
```

```python
data_pred=pd.DataFrame({'YearsExperience':[2,3,4,5,6,7,8,9,10]})
data_pred
```
